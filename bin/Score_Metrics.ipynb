```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
data = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]
data = pd.read_csv("/home/rodtheo/Downloads/assemblies_table-plot(1).tsv", sep="\t", header=0, index_col=False)
data
```

```python
greater_better = ['Merfin QV*', 'Error-free bases (%)', 'N50', 'BUSCO complete (%)', 'BUSCO single (%)']
```

```python
smaller_better = ['REAPR erros', 'BUSCO duplicated (%)', 'BUSCO fragmented (%)', 'BUSCO missing (%)']
complement = data[smaller_better].max(axis=0)
complement
```

```python
data[smaller_better]
```

```python
complement
```

```python
complement.sub(data[smaller_better])
```

```python
scaler_sb = MinMaxScaler()
subset_sb = complement - data[smaller_better]
scaler_sb.fit(subset_sb)
scaler_sb_df = scaler_sb.transform(subset_sb)
scaler_sb_df
```

```python
scaler = MinMaxScaler()
scaler.fit(data[greater_better])
scaler_df = scaler.transform(data[greater_better])
```

```python
score = np.mean(np.concatenate((scaler_df, scaler_sb_df), axis=1), axis=1)
score
```

```python
data['Score'] = score
```
